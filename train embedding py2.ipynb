```python
train = pd.read_csv('./data/SST2/train.csv')
test = pd.read_csv('./data/SST2/test.csv')
dev = pd.read_csv('./data/SST2/dev.csv')
```

```python
train.shape,test.shape,dev.shape
```

```python
text = list(train['text'])
len(text)
```

```python
text += list(test['text'])
text += list(dev['text'])
len(text)
```

```python
sentences = [i.split() for i in text if 'https' not in i]
```

```python
sentences
```

```python
from gensim.models import Word2Vec
```

```python
import time
```

```python
start = time.time()
model = Word2Vec(sentences, size=300, window=6, min_count=5, workers=4)
end = time.time()
```

```python
print "the transformation takes {0:.2f} minutes".format((end-start)/60 )
```

```python
model.wv['emoji1']
```

```python
all_embed = model.wv
```
